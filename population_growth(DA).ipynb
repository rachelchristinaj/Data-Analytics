# POPULATION GROWTH RATE IN 2024

```python
plt.figure(figsize=(12,4))
ax=sns.barplot(x='Population (2024)' , y='Growth Rate' , data=df)
plt.show()
```

# FREQUENCY OF POPULATION IN 2023

```python
plt.figure(figsize=(15, 5))
sns.histplot(df['Population (2023)'], kde=True, bins=30)
plt.title('Distribution of Population in 2023')
plt.xlabel('Population (2023)')
plt.ylabel('Frequency')
plt.show()
```

# FREQUENCY OF POPULATION IN 2024

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='Growth Rate', y='Population (2024)', data=df)
plt.title('Population  growth rate in 2024')
plt.xlabel('Growth Rate')
plt.ylabel('Population (2024)')
plt.xticks(rotation=45)
plt.show()
```

# Total Count in Continent-Wise

```python
plt.figure(figsize=(12, 6))
ax = sns.countplot(data = df , x = "Continent")
ax.bar_label(ax.containers[0])
plt.show()
```
